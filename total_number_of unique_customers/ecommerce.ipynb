```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

ecom = pd.read_csv('d:\\Ecommerce Analytics\\files\\ecommerce_data.csv')
```

```python
ecom['Quantity'] = ecom['Quantity'].fillna(0)
ecom['PurchaseID'] = ecom['PurchaseID'].fillna(0)
ecom['PurchaseDate'] = pd.to_datetime(ecom['PurchaseDate'])

ecom
```

```python
ecom['FullName'] = ecom['FirstName'] + ' ' + ecom['LastName']

ecom
```

```python
for id in ecom['PurchaseID']:
    if id > 0:
        unique_cust_sum = ecom['FullName'].unique().size


print(f"Total number of total unique customers with greater than 0 purchases: {unique_cust_sum}")
```
