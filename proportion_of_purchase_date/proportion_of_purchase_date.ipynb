```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

ecom = pd.read_csv('files/ecommerce_data.csv')
```

```python
ecom['Quantity'] = ecom['Quantity'].fillna(0)
ecom['PurchaseID'] = ecom['PurchaseID'].fillna(0)
ecom['PurchaseDate'] = pd.to_datetime(ecom['PurchaseDate'])
ecom['RegistrationDate'] = pd.to_datetime(ecom['RegistrationDate'])
ecom['FullName'] = ecom['FirstName'] + ' ' + ecom['LastName']

ecom = ecom.drop_duplicates(subset=['FullName'], keep='first')
ecom['Age'] = ecom['Age'].fillna(0)

ecom['Age'] = ecom['Age'].astype('int')

ecom 
```

```python
count = 0

for i, (PurchaseDate, RegistrationDate) in enumerate(zip(ecom['PurchaseDate'], ecom['RegistrationDate'])):
    if PurchaseDate < RegistrationDate:
        count += 1

proportion = count / len(ecom['RegistrationDate'])
print(f"The proportion of customers who made a purchase before registering is: {proportion:.2%}")
```
