```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import re
from datetime import datetime
```

```python
ecom = pd.read_csv('d:\\Ecommerce Analytics\\files\\ecommerce_data.csv')
```

```python
ecom["PurchaseDate"]    = pd.to_datetime(ecom["PurchaseDate"],    errors="coerce")
ecom["RegistrationDate"]= pd.to_datetime(ecom["RegistrationDate"],errors="coerce")
```

```python
state_avg = ecom.groupby("State")["TotalAmount"].mean().sort_values(ascending=False)
print(f"State with highest avg completed TotalAmount: {state_avg.idxmax()} (${state_avg.max():.2f})")
```
