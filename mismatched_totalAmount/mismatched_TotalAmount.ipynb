```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import re
from datetime import datetime

ecom = pd.read_csv('d:\\Ecommerce Analytics\\files\\ecommerce_data.csv')
```

```python
ecom["PurchaseDate"]    = pd.to_datetime(ecom["PurchaseDate"],    errors="coerce")
ecom["RegistrationDate"]= pd.to_datetime(ecom["RegistrationDate"],errors="coerce")
```

```python
ecom["ExpectedTotal"]  = ecom["Price"] * ecom["Quantity"]
ecom["AmountMismatch"] = ~np.isclose(ecom["TotalAmount"], ecom["ExpectedTotal"], atol=0.01)
mismatch_count = ecom["AmountMismatch"].sum()
print(f"Purchases with amount mismatch: {mismatch_count}")
```
